```python
text = 'Hello World, this is a simple example to show how tokenization works in NLP and an sentence'
```

# **TOKENIZATION PROCESS**

```python
tokens = text.split()

print("tokens:", tokens)
```

```python
vocab = {}
current_id = 0

for token in tokens:
    if token not in vocab:
        vocab[token] = current_id
        current_id += 1

print("vocab:", vocab)
```

## **ENCODE / DECODE**

```python
# ENCODE
encoded = [vocab[token] for token in tokens]
print(encoded)
```

```python
# DECODE
reverse_vocab = {id: token for token, id in vocab.items()}
# print("reverse_vocab:", reverse_vocab)

decoded = " ".join([reverse_vocab[id] for id in encoded])
print("decoded:", decoded)
```

# **EMBEDING VEKTOR**

```python
import torch
import torch.nn as nn

# dim tanlovi parametrlarni belgilaydi
# tiny project (8–32) & small NLP (64–128) real NLP (256–300) & LLM (768+)
# parameters = vocab_size × embedding_dim > koproq parametr organadi
embedding_dim = 3
vocab_size = len(vocab)

embedding_layer = nn.Embedding(vocab_size, embedding_dim)

input_ids = torch.tensor([encoded])
embeddings = embedding_layer(input_ids)

print(embeddings)
print(embeddings.shape)
```
