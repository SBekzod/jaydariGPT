```python
text = 'Hello World, this is a simple example to show how tokenization works in NLP and a sentence'
```

# **TOKENIZATION PROCESS**

```python
tokens = text.split()

print("tokens:", tokens)
```

```python
vocab = {}
current_id = 0

for token in tokens:
    if token not in vocab:
        vocab[token] = current_id
        current_id += 1

print("vocab:", vocab)
```

## **ENCODE / DECODE**

```python
# ENCODE
encoded = [vocab[token] for token in tokens]
print(encoded)
```

```python
# DECODE
reverse_vocab = {id: token for token, id in vocab.items()}
# print("reverse_vocab:", reverse_vocab)

decoded = " ".join([reverse_vocab[id] for id in encoded])
print("decoded:", decoded)
```

# **EMBEDING VEKTOR**

```python
import torch
import torch.nn as nn

# dim tanlovi parametrlarni belgilaydi
# tiny project (8–32) & small NLP (64–128) real NLP (256–300) & LLM (768+)
# parameters = vocab_size × embedding_dim > koproq parametr organadi
embedding_dim = 3
vocab_size = len(vocab)

embedding_layer = nn.Embedding(vocab_size, embedding_dim)

input_ids = torch.tensor([encoded])
embeddings = embedding_layer(input_ids)

print(embeddings)
print(embeddings.shape)
```

```python
from transformers import AutoTokenizer

model_id = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"
tokenizer = AutoTokenizer.from_pretrained(model_id)
```

```python
print(tokenizer.vocab_size)
print(tokenizer.special_tokens_map) # BOS EOS UNK PAD
```

```python
text = 'Hello how are you doing?'
print('text:', text)
```

```python
tokens = tokenizer.tokenize(text)
print("tokens:", tokens, "\n")

ids = tokenizer.convert_tokens_to_ids(tokens)
print("TokenIds:", ids, "\n")

encoded = tokenizer(text)
print("Encoding:", encoded, "\n")
# <BOS> Hello how are you doing ?
# 1 bu tokenni o‘qish kerakligi, 0 esa bu padding yani etibor bermang
```

```python
# pad_token - faked value
# padding - 0 mask

"Hi" #=> [1, 345, 2, 2, 2] // length 5
"How are you?" #=> [1, 45, 657, 234, 23453] // length 5
# attention_mask: [1, 1, 0, 0, 0]
```

```python
decoded_text = tokenizer.decode(encoded["input_ids"])
print('decoded_text:', decoded_text)
```
